```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
```

```python
np.random.seed(42)
X = np.random.choice(['a', 'b', 'c'], size=1000)
y = np.arange(1000)
df = pd.DataFrame()
df['feature'] = X
df['target'] = y
df
```

```python
def mean_encode(split, feature, target, alpha, dtype=np.object):
    train_id, test_id = split
    X_train, y_train = feature[train_id], target[train_id]
    res = np.zeros_like(feature, dtype=dtype)
    
    for value in set(X_train):
        n_rows = len(X_train[X_train == value])
        mean = np.mean(y_train[X_train == value])
        global_mean = np.mean(y_train)
        res[feature == value] = mean_encoding = (mean * n_rows + global_mean * alpha) / (n_rows + alpha)
    
    return res
```

```python
for i, split in enumerate(KFold(n_splits=5, random_state=42).split(X)):
    df[f'split_{i}'] = mean_encode(split, X, y, 0.5)
```
